```python
import numpy as np
import torch
import string

def get_preds(text):
    max_proba = get_proba(text)[0].argsort()[:,-3]#np.argmax(get_proba(text))
    return train_vocab.label_encoder.decode(max_proba)

def get_proba(text):
    if isinstance(text, str):
        text = [txt for txt in text.lower().translate(str.maketrans('', '', string.punctuation)).split(' ') if txt not in '']
        
    str_enc = [torch.as_tensor(train_vocab.label_encoder.encode(txt)) for txt in text]
    output = model(([torch.stack(str_enc)], torch.stack([torch.as_tensor(len(str_enc)).long().reshape(1,)])))
    
    sm = torch.nn.Softmax(dim=1)
#     print([sm(out).detach().cpu().numpy() for out in [output]])
    return [sm(out).detach().cpu().numpy() for out in [output]]
```
