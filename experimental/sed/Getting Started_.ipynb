#### 3.1.1 Data preprocessing and preparation

##### Prepare sentiment/emotion data to pass into the model

```python
from pathlib import Path

DATA_DIR='./emotions'#path to csv data file
wd = Path(DATA_DIR)

emotions = ['angry', 'fear', 'happy', 'sad']
sentiments = ['positive', 'negative']
```

```python
import pandas as pd

def merge_df(working_dir: str, emotions):
    """
    Helper function to prepare the data by adding appropriate labels
    """

    dfs = []
    for cl in emotions:
        df = pd.read_csv(working_dir.joinpath(f'{cl}.csv')).drop(columns=['Unnamed: 1']).dropna()
        df['emotion'] = cl
        if cl == 'happy':
            df['sentiment'] = 'positive'
        else:
            df['sentiment'] = 'negative'
        dfs.append(df)
        
    return pd.concat(dfs)
```

```python
from sklearn.preprocessing import LabelEncoder

#load and prepare dataframe
df = merge_df(wd, emotions)
df = df[df['Sentensi'].str.split().str.len().ge(2)]

#encoding the labels
sle, ele = LabelEncoder(),LabelEncoder()
for col,le in zip(['emotion', 'sentiment'], [ele, sle]):
    df[col] = le.fit_transform(df[col])
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

#split data into train-test sets
sss = StratifiedShuffleSplit(n_splits=1, test_size=0.3, random_state=1947)

for train_index, val_index in sss.split(df.index, df.emotion):
    break
    
train_df, test_df = df.iloc[train_index, :].reset_index(drop=True), df.iloc[val_index, :].reset_index(drop=True)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

#split training data into train-validation sets
sss = StratifiedShuffleSplit(n_splits=1, test_size=0.3, random_state=1947)

for train_index, val_index in sss.split(train_df.index, train_df.emotion):
    break
    
train, val = train_df.iloc[train_index, :].reset_index(drop=True), train_df.iloc[val_index, :].reset_index(drop=True)
```

```python
import string

#extract text data from dataframe
strings = [t for txt in df['Sentensi'].values.tolist() for t in txt.lower().translate(str.maketrans('', '', string.punctuation)).split(' ') if t not in '']

train_text = [list(filter(None, txt.lower().translate(str.maketrans('', '', string.punctuation)).split(' '))) for txt in train['Sentensi'].values.tolist()]
val_text = [list(filter(None, txt.lower().translate(str.maketrans('', '', string.punctuation)).split(' '))) for txt in val['Sentensi'].values.tolist()]
test_text = [list(filter(None, txt.lower().translate(str.maketrans('', '', string.punctuation)).split(' '))) for txt in test_df['Sentensi'].values.tolist()]
```

```python
import numpy as np
import torch
import string

def get_preds(probs, eps=0.05):
    class_dict, overalls = {}, {}
    
    for prob, le, cl in zip(probs ,[ele, sle], ['emotion', 'sentiment']):
        class_dict[f'{cl}_distribution'] = {le.inverse_transform([i])[0] : prob[0,i] for i in range(prob.shape[1])}
        
        if cl != 'sentiment' or (cl == 'sentiment' and abs(prob[0,0]-prob[0,1])>=eps):
            overalls[f'overall_{cl}'] = le.inverse_transform([np.argmax(prob)])[0]
        else:
            overalls[f'overall_{cl}'] = 'neutral'
    
    return class_dict, overalls

def get_proba(text):
    if isinstance(text, str):
        text = [txt for txt in text.lower().translate(str.maketrans('', '', string.punctuation)).split(' ') if txt not in '']

    str_enc = tokenizer.transform(text)
    
    output = model(([str_enc], [[None for j in range(len(str_enc))]]))
    sm = torch.nn.Softmax(dim=1)
    return [sm(out) for out in output]
    

def get_analysis(text):
    probs = get_proba(text)
    return get_preds(probs)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report

def analyze_predictions(labels, preds):
    print(classification_report(labels, preds))
    print(f1_score(labels, preds, average='weighted'))
    print(accuracy_score(labels, preds))
```
